### Q1. ë„’ì´ ê³„ì‚°ê¸°
- ë„í˜•ì˜ ëª¨ì–‘ê³¼ ì •ë³´ë¥¼ ì…ë ¥ë°›ì•„ ë„“ì´ë¥¼ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.
- ë§¤ê°œë³€ìˆ˜ : width = 0, height = 0, radius = 0, pi = 3.14, shape = '' (ëª¨ë‘ ë””í´íŠ¸ ë§¤ê°œë³€ìˆ˜ë¡œ ì„¤ì •)
- shapeê°€ 'circle' ì´ë©´ ì›ì˜ ë„“ì´, 'triangle' ì´ë©´ ì‚¼ê°í˜•ì˜ ë„“ì´, 'rectangle' ì´ë©´ ì‚¬ê°í˜•ì˜ ë„“ì´ë¥¼ ì¶œë ¥í•˜ì„¸ìš”.

```python
# ìœ„ì— ì •ì˜í•œ ë‚´ìš©ì„ ì°¸ê³ í•˜ì—¬ getArea() í•¨ìˆ˜ë¥¼ ì •ì˜í•˜ì„¸ìš”

def getArea(width = 0, height = 0, radius = 0, pi = 3.14, shape = ''):
  
  if shape == "triangle": #ì‚¼ê°í˜•
    area = width*height*(1/2)
    return area

  if shape == "rectangle": #ì‚¬ê°í˜•
    area = width*height
    return area

  if shape == "circle": #ì›
    area = radius*radius*pi
    return area

```

```python
# width 10, height 20 ì¸ ì§ì‚¬ê°í˜•ì˜ ë„“ì´ë¥¼ êµ¬í•˜ì„¸ìš”.
print(getArea(10, 20, shape = 'rectangle')) #ì¸ì ì—†ì´ ë„˜ì–´ì˜¤ë©´, ìœ„ì¹˜ ìˆœì„œì— ë”°ë¼!

# width 10, height 20 ì¸ ì‚¼ê°í˜•ì˜ ë„“ì´ë¥¼ êµ¬í•˜ì„¸ìš”.
print(getArea(10, 20, shape = 'triangle'))

# radius 10 ì¸ ì›ì˜ ë„“ì´ë¥¼ êµ¬í•˜ì„¸ìš”.
print(getArea(radius = 10, shape = 'circle'))

# radius 10, pi 3.141592 ì¸ ì›ì˜ ë„“ì´ë¥¼ êµ¬í•˜ì„¸ìš”.
print(getArea(radius = 10, pi = 3.141592, shape = 'circle'))
```

### Q2. í•™ì  ê³„ì‚°ê¸°
- í•™ì ì„ ì…ë ¥ë°›ì•„ í‰ê·  ì ìˆ˜ë¥¼ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.
- ë§¤ê°œë³€ìˆ˜ : grades (í•™ì ì´ ','ë¡œ êµ¬ë¶„ë˜ì–´ìˆëŠ” ë¬¸ìì—´)
    - ex) "A+,B,D,B+,C"
- ë‹¤ìŒ ì ìˆ˜ ë³€í™˜ ê·œì¹™ì— ë”°ë¼ ê³„ì‚°í•˜ì—¬ í‰ê·  ì ìˆ˜ë¥¼ ë¦¬í„´
    - 'A+' : 4.5, 'A' : 4.0, 'B+' : 3.5, 'B': 3.0, 'C' : 2.0, 'D' : 1.0

```python
# getMeanScore() í•¨ìˆ˜ë¥¼ ì •ì˜í•˜ì„¸ìš”
import numpy as np


def getMeanScore (grades):
  grades = grades.split(',')
  arr = [] 

  for i in grades :
    if i == 'A+' :
      arr.append(4.5)

    if i == 'A' :
      arr.append(4.0)

    if i == 'B+' :
      arr.append(3.5)

    if i == 'B' :
      arr.append(3.0)

    if i == 'C' :
      arr.append(2.0)

    if i == 'D' :
      arr.append(1.0)                      

  return np.mean(arr)





```

```python
# ì½”ë“œë¥¼ ì‹¤í–‰í•˜ì„¸ìš”.
grades = input('í•™ì ì„ ì…ë ¥í•˜ì„¸ìš”.(ex. A+,B,D,B+,C) : ')
getMeanScore(grades)
```
